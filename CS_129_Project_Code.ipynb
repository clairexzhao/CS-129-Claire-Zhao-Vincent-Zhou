```python
#1. Imports
import pandas as pd
import numpy as np
import re
from collections import Counter

import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.metrics import (
    accuracy_score,
    precision_recall_fscore_support,
    classification_report,
    confusion_matrix,
    ConfusionMatrixDisplay
)
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

# 2. Load Dataset
import kagglehub
path = kagglehub.dataset_download("elvinrustam/wine-dataset")
import os
print(path)
print(os.listdir(path))
file_path = os.path.join(path, "WineDataset.csv")
df = pd.read_csv(file_path)

print("Dataset shape:", df.shape)
print("\nColumns:")
print(df.columns.tolist())
print("\nFirst 5 rows:")
display(df.head())

#3. Build flavour dictionary using characteristics and returns label characteristics
flavor_map = {
    "Fruity": [
        "apple", "green apple", "red apple", "pear", "peach", "apricot", "nectarine",
        "plum", "black plum", "red plum", "berry", "blackberry", "blackcurrant",
        "blueberry", "raspberry", "strawberry", "cranberry", "cherry", "black cherry",
        "red cherry", "citrus fruit", "citrus", "lemon", "lime", "grapefruit",
        "orange", "melon", "passionfruit", "pineapple", "tropical fruit",
        "gooseberry", "stone fruit", "quince", "fig", "raisin", "jammy",
        "red fruit", "black fruit", "ripe fruit"
    ],
    "Oaky": [
        "vanilla", "oak", "cedar", "toast", "toasty", "smoke", "smoky",
        "bread", "biscuit", "pastry", "cream", "butter", "coconut", "coffee",
        "chocolate", "tobacco"
    ],
    "Spicy": [
        "spice", "spicy", "black pepper", "white pepper", "pepper", "cloves",
        "clove", "nutmeg", "sweet spice", "cinnamon", "anise"
    ],
    "Earthy": [
        "earth", "earthy", "mineral", "wet stones", "stone", "mushroom",
        "leather", "graphite", "savoury", "petrol"
    ],
    "Floral": [
        "floral", "blossom", "honeysuckle", "violet", "rose", "jasmine"
    ],
    "Herbal": [
        "grass", "herbaceous", "dried herbs", "mint", "fennel", "sage",
        "thyme", "green bell pepper", "eucalyptus", "herbal"
    ]
}

#label function
def assign_primary_flavor(characteristics, flavor_map):

    if pd.isna(characteristics):
        return np.nan, {}

    descriptors = [x.strip().lower() for x in characteristics.split(",")]

    for desc in descriptors:

        for category, keywords in flavor_map.items():

            for kw in keywords:

                if kw in desc:
                    return category, {category:1}

    return np.nan, {}

    descriptors = [clean_descriptor(x) for x in str(characteristics).split(",")]
    counts = Counter()

    for desc in descriptors:
        for category, keywords in flavor_map.items():
            for kw in keywords:
                # exact phrase containment
                if kw in desc:
                    counts[category] += 1

    if len(counts) == 0:
        return np.nan, {}

    max_count = max(counts.values())
    top_categories = [k for k, v in counts.items() if v == max_count]

    # safest for clean single-label classification
    if len(top_categories) > 1:
        return np.nan, dict(counts)

    return top_categories[0], dict(counts)

results = df["Characteristics"].apply(lambda x: assign_primary_flavor(x, flavor_map))
df["label"] = results.apply(lambda x: x[0])
df["label_counts"] = results.apply(lambda x: x[1])

print("Label distribution before dropping unlabeled rows:")
print(df["label"].value_counts(dropna=False))

#Drop rows with no clean primary label
df = df.dropna(subset=["label"]).copy()

print("\nDataset shape after label construction:", df.shape)
print("\nLabel distribution after dropping unlabeled rows:")
print(df["label"].value_counts())

#5. Keep only columns we need (features that are relevant to taste and also easily observable from the packaging)
feature_cols = [
    "Price",
    "Grape",
    "Country",
    "Type",
    "ABV",
    "Region",
    "Vintage",
    "Appellation"
]

df_model = df[feature_cols + ["Characteristics", "label"]].copy()
display(df_model.head())

#6. Clean numeric columns
# Price looks like: "£15.99 per bottle"
df_model["Price"] = (
    df_model["Price"]
    .astype(str)
    .str.extract(r"([0-9]+(?:\.[0-9]+)?)")[0]
)
df_model["Price"] = pd.to_numeric(df_model["Price"], errors="coerce")

# ABV looks like: "ABV 13.50%"

df_model["ABV"] = (
    df_model["ABV"]
    .astype(str)
    .str.extract(r"([0-9]+(?:\.[0-9]+)?)")[0]
)

df_model["ABV"] = pd.to_numeric(df_model["ABV"], errors="coerce")

# Vintage looks like either a year (e.g. 2021) or "NV"
# "NV" means non-vintage, so we convert it to missing
df_model["Vintage"] = df_model["Vintage"].astype(str).str.strip()
df_model["Vintage"] = df_model["Vintage"].replace("NV", np.nan)
df_model["Vintage"] = pd.to_numeric(df_model["Vintage"], errors="coerce")

print(df_model[["Price", "ABV", "Vintage"]].dtypes)
print(df_model[["Price", "ABV", "Vintage"]].head(10))
print(df_model[["Price", "ABV", "Vintage"]].isnull().sum())
print(df_model["Type"].value_counts())
print(df_model["label"].value_counts(dropna=False))

#7. Collapse rare categories into "Other"
categorical_cols = ["Grape", "Country", "Type", "Region", "Appellation"]
numeric_cols = ["Price", "ABV", "Vintage"]

def collapse_rare_categories(series, min_count=10):
    counts = series.value_counts(dropna=False)
    keep = counts[counts >= min_count].index
    return series.where(series.isin(keep), other="Other")

for col in categorical_cols:
    df_model[col] = df_model[col].astype(str)
    df_model[col] = collapse_rare_categories(df_model[col], min_count=10)
#sanity check
for col in categorical_cols:
    print(f"\n{col} top values:")
    print(df_model[col].value_counts().head(10))

#7. Train / validation / test split
from sklearn.model_selection import train_test_split
X = df_model[feature_cols].copy()
y = df_model["label"].copy()

X_temp, X_test, y_temp, y_test = train_test_split(
    X, y,
    test_size=0.15,
    stratify=y,
    random_state=42
)

val_size_relative = 0.15 / 0.85

X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp,
    test_size=val_size_relative,
    stratify=y_temp,
    random_state=42
)

print("Train size:", len(X_train))
print("Validation size:", len(X_val))
print("Test size:", len(X_test))

#Check class imbalance
print("\nTrain label proportions:")
print(y_train.value_counts(normalize=True).round(3))

print("\nValidation label proportions:")
print(y_val.value_counts(normalize=True).round(3))

print("\nTest label proportions:")
print(y_test.value_counts(normalize=True).round(3))

#8. Build preprocessing pipeline
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler

preprocessor = ColumnTransformer(
    transformers=[
        ("cat", Pipeline([
            ("imputer", SimpleImputer(strategy="most_frequent")),
            ("onehot", OneHotEncoder(handle_unknown="ignore"))
        ]), categorical_cols),
        ("num", Pipeline([
            ("imputer", SimpleImputer(strategy="median")),
            ("scaler", StandardScaler())
        ]), numeric_cols)
    ]
)

#9. Evaluation helper
from sklearn.metrics import (
    accuracy_score,
    precision_recall_fscore_support,
    classification_report,
    confusion_matrix,
    ConfusionMatrixDisplay
)
import matplotlib.pyplot as plt

def evaluate_model(y_true, y_pred, model_name="model"):
    acc = accuracy_score(y_true, y_pred)

    precision_macro, recall_macro, f1_macro, _ = precision_recall_fscore_support(
        y_true, y_pred, average="macro", zero_division=0
    )

    precision_weighted, recall_weighted, f1_weighted, _ = precision_recall_fscore_support(
        y_true, y_pred, average="weighted", zero_division=0
    )

    print(f"\n==============================")
    print(model_name)
    print("==============================")
    print(f"Accuracy:        {acc:.3f}")
    print(f"Macro Precision: {precision_macro:.3f}")
    print(f"Macro Recall:    {recall_macro:.3f}")
    print(f"Macro F1:        {f1_macro:.3f}")
    print(f"Weighted F1:     {f1_weighted:.3f}")
    print("\nClassification report:")
    print(classification_report(y_true, y_pred, zero_division=0))

    return {
        "model": model_name,
        "accuracy": acc,
        "macro_precision": precision_macro,
        "macro_recall": recall_macro,
        "macro_f1": f1_macro,
        "weighted_f1": f1_weighted
    }

#10. Majority class baseline
results_summary = []

majority_class = y_train.value_counts().idxmax()
y_val_pred_majority = np.repeat(majority_class, len(y_val))

results_summary.append(
    evaluate_model(y_val, y_val_pred_majority, model_name="Majority-Class Baseline")
)

#11. Logistic Regression with Tuning
from sklearn.linear_model import LogisticRegression

C_values = [0.01, 0.1, 1.0, 10.0]
best_logreg_model = None
best_logreg_f1 = -1
best_C = None

print("Tuning Logistic Regression...")

for C in C_values:
    model = Pipeline([
        ("preprocessor", preprocessor),
        ("clf", LogisticRegression(
            C=C,
            multi_class="multinomial",
            solver="lbfgs",
            max_iter=3000
        ))
    ])

    model.fit(X_train, y_train)
    y_val_pred = model.predict(X_val)

    macro_f1 = precision_recall_fscore_support(
        y_val, y_val_pred, average="macro", zero_division=0
    )[2]

    print(f"C = {C:<5} | Validation Macro F1 = {macro_f1:.3f}")

    if macro_f1 > best_logreg_f1:
        best_logreg_f1 = macro_f1
        best_logreg_model = model
        best_C = C

print("\nBest C:", best_C)

y_val_pred_logreg = best_logreg_model.predict(X_val)

results_summary.append(
    evaluate_model(y_val, y_val_pred_logreg, model_name=f"Best Logistic Regression (C={best_C})")
)

#12. Random Forest
from sklearn.ensemble import RandomForestClassifier

rf_model = Pipeline([
    ("preprocessor", preprocessor),
    ("clf", RandomForestClassifier(
        n_estimators=100,
        random_state=42
    ))
])

rf_model.fit(X_train, y_train)
y_val_pred_rf = rf_model.predict(X_val)

results_summary.append(
    evaluate_model(y_val, y_val_pred_rf, model_name="Random Forest (n_estimators=100)")
)

#13. Neural Networks
from sklearn.neural_network import MLPClassifier

mlp_model = Pipeline([
    ("preprocessor", preprocessor),
    ("clf", MLPClassifier(
        hidden_layer_sizes=(128, 64),
        activation="relu",
        solver="adam",
        alpha=0.0001,
        learning_rate_init=0.001,
        max_iter=500,
        random_state=42
    ))
])

mlp_model.fit(X_train, y_train)
y_val_pred_mlp = mlp_model.predict(X_val)

results_summary.append(
    evaluate_model(y_val, y_val_pred_mlp, model_name="Neural Network (MLP)")
)

#14. Compare models
results_df = pd.DataFrame(results_summary).sort_values("macro_f1", ascending=False)
display(results_df)

#15. Confusion matrix for best validation model
best_model_name = results_df.iloc[0]["model"]

if "Logistic Regression" in best_model_name:
    best_val_model = best_logreg_model
    best_val_preds = y_val_pred_logreg

elif "Random Forest" in best_model_name:
    best_val_model = rf_model
    best_val_preds = y_val_pred_rf

elif "Neural Network" in best_model_name or "MLP" in best_model_name:
    best_val_model = mlp_model
    best_val_preds = y_val_pred_mlp

else:
    raise ValueError(f"Unknown best model: {best_model_name}")

labels_sorted = sorted(y_val.unique())

cm = confusion_matrix(y_val, best_val_preds, labels=labels_sorted)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels_sorted)

fig, ax = plt.subplots(figsize=(8, 8))
disp.plot(ax=ax, xticks_rotation=45, colorbar=False)
plt.title(f"Validation Confusion Matrix: {best_model_name}")
plt.tight_layout()
plt.show()

#16. Per-class metrics
report_dict = classification_report(
    y_val,
    best_val_preds,
    output_dict=True,
    zero_division=0
)

report_df = pd.DataFrame(report_dict).T
display(report_df)

#17. Error Analysis
X_val_reset = X_val.reset_index()
y_val_reset = y_val.reset_index(drop=True)

error_df = X_val_reset.copy()
error_df["true_label"] = y_val_reset
error_df["pred_label"] = best_val_preds

error_df = error_df.merge(
    df[["Characteristics"]].reset_index(),
    on="index",
    how="left"
)

misclassified = error_df[error_df["true_label"] != error_df["pred_label"]].copy()

print("Number of misclassified validation examples:", misclassified.shape[0])

display(
    misclassified[
        ["true_label", "pred_label", "Grape", "Country", "Type", "ABV", "Region",
         "Vintage", "Appellation", "Characteristics"]
    ].head(20)
)

#18. Final test evaluation - retrain the best model on train and validation and then evaluate once on test
X_trainval = pd.concat([X_train, X_val], axis=0)
y_trainval = pd.concat([y_train, y_val], axis=0)

if "Logistic Regression" in best_model_name:
    final_model = Pipeline([
        ("preprocessor", preprocessor),
        ("clf", LogisticRegression(
            C=best_C,
            solver="lbfgs",
            max_iter=3000
        ))
    ])

elif "Random Forest" in best_model_name:
    final_model = Pipeline([
        ("preprocessor", preprocessor),
        ("clf", RandomForestClassifier(
            n_estimators=100,
            random_state=42
        ))
    ])

elif "Neural Network" in best_model_name or "MLP" in best_model_name:
    final_model = Pipeline([
        ("preprocessor", preprocessor),
        ("clf", MLPClassifier(
            hidden_layer_sizes=(128, 64),
            activation="relu",
            solver="adam",
            alpha=0.0001,
            learning_rate_init=0.001,
            max_iter=500,
            random_state=42
        ))
    ])

else:
    raise ValueError(f"Unknown best model: {best_model_name}")

final_model.fit(X_trainval, y_trainval)
y_test_pred = final_model.predict(X_test)

final_test_metrics = evaluate_model(
    y_test,
    y_test_pred,
    model_name=f"Final Test Performance ({best_model_name})"
)

#19. Test confusion matrix
cm_test = confusion_matrix(y_test, y_test_pred, labels=labels_sorted)
disp_test = ConfusionMatrixDisplay(confusion_matrix=cm_test, display_labels=labels_sorted)

fig, ax = plt.subplots(figsize=(8, 8))
disp_test.plot(ax=ax, xticks_rotation=45, colorbar=False)
plt.title("Test Confusion Matrix")
plt.tight_layout()
plt.show()


```
